```python
import numpy as np
import matplotlib.pyplot as plt

Fs = 5000
n = np.arange(0, 100)
Fo_list = [0.5e3, 2e3, 3e3, 4.5e3]

for Fo in Fo_list:
  Xn = np.sin(2 * np.pi * (Fo / Fs) * n)
  plt.figure(figsize=(10, 4))
  plt.plot(n, Xn)
  plt.title('Plot of X(n) for F0 = {} kHz'.format(Fo / 1e3))
  plt.xlabel('n')
  plt.ylabel('X(n)')
  plt.grid(True)
  plt.show()
```

```python
Fo = 2000
Fs = 50000

Xn = np.sin(2 * np.pi * (Fo / Fs) * n)
plt.figure(figsize=(10, 4))
plt.plot(n, Xn)
plt.title('Plot of X(n) for F0 = {} kHz'.format(Fo / 1e3))
plt.xlabel('n')
plt.ylabel('X(n)')
plt.grid(True)
plt.show()
```

```python
# Find indices where Yn has values
valid_indices = np.where(Yn != 0)[0]

# Slice n and Yn to only include overlapping range
n_valid = n[valid_indices]
Yn_valid = Yn[valid_indices]

# Plot the overlapping range
plt.figure(figsize=(10, 4))
plt.plot(n_valid, Yn_valid)
plt.title('Plot of X(n) for F0 = {} kHz'.format(Fo / 1e3))
plt.xlabel('n')
plt.ylabel('Y(n)')
plt.grid(True)
plt.show()
```
